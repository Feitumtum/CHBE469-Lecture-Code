# 2) Subplots
We will use subplots to put multiple plots and organize them in the figure

(1) We will need to specify the location of the plot before we plot them.
* plt.subplot(nrows, ncols, plot_number)

(2) nrows and ncols specify the the number of spots in your figure
* 2 x 3 = 6 plots (2 rows of 3 graphs each)

(3) The plot_number specify which plot you want to plot on

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'r:')
```

# 3) Titles, Labels, Legends, Figure Size, Exporting

## We can add title for each graph and axes based on the subplot
* Notice how the title and axis labels for one graph is done before we specify the next graph by using plt.subplot(1,2,2)
* You have to specify everything for one plot before moving on the the next plot

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.title("PLOT 1")
plt.ylabel("Power")
plt.xlabel("RPM")

plt.subplot(1,2,2)
plt.plot(y, x, 'r:')
plt.title("PLOT 2")
plt.xlabel("RPM")
```

# 6) Object Oriented Approach to Plotting
Previously, we saw a very easy way to plot quickly in Python if we want to just use the methods from plt. Another method of plotting is to actually work with the individual components and then call functions on them.

## Example 1 (Single Plot)

```python
fig = plt.figure()

# Then we can add an axes component to the figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Then we can add a plot to the axes
axes.plot(x,y)

# Next, we can label our graphs with axes titles and a graph title
# Notice how it is set_xlabel instead of xlabel
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Graph Title')
```

## Example 2 (Subplots)

```python
fig, axes = plt.subplots(nrows=1, ncols=5)
fig.set_figwidth(20) 
```

Now that we have the fig and axes components, we can use a for loop to plot on each plot in a systematic way

```python
# Using tuple unpacking we extra the fig and axes components from the plt.subplot()
fig, axes = plt.subplots(nrows=1, ncols=5)
fig.set_figwidth(20) 

for i, ax in enumerate(axes):
    ax.plot(x, y**i/2, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

Alternatively if we want a different title/labels based on each graph we can loop through a list as well

```python
titles = ["Title 1", "Title 2", "Title 3", "Title 4", "Title 5"]
x_axis = ["X1", "X2", "X3", "X4", "X5"]
y_axis = ["Y1", "Y2", "Y3", "Y4", "Y5"]

fig, axes = plt.subplots(nrows=1, ncols=5)
fig.set_figwidth(20) 

for i, ax in enumerate(axes):
    ax.plot(x, y**i/2, 'b')
    ax.set_xlabel(x_axis[i]) # accessing list based on iteration
    ax.set_ylabel(y_axis[i])
    ax.set_title(titles[i])
    
fig.tight_layout()
```

## We can have a secondary y axis

```python
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Common X')
ax1.set_ylabel('Y1', color = color)
ax1.plot(x, x**2, color = color)
ax1.tick_params(axis ='y', labelcolor = color)

ax2 = ax1.twinx()

color = 'tab:green'
ax2.set_ylabel('Y2', color = color)
ax2.plot(x, 1/x, color = color)
ax2.tick_params(axis ='y', labelcolor = color)

plt.title('TITLE')
```

### All the different things we learned before about figure size, legend, labeling, etc can all be used in the object oriented method as well but using slighlty different method names.

Examples:
* ylabel --> set_ylabel
* title --> set_title

It's also important to notice how if we have multiple plots, we will have to specify which one by using `ax[i]`

```python
fig, ax = plt.subplots(nrows=1, ncols=2, dpi = 75)
fig.set_size_inches(16,8)

ax[0].plot(x, x**2, 'k--o')
ax[0].set_xlabel('X1 Label')
ax[0].set_ylabel('Y1 Label')
ax[0].set_title('Graph Title 1')
ax[0].text(7, 40, r"$y=x^2$", fontsize=20, color="k")

ax[1].plot(x, x**(2/3), 'r--o')
ax[1].set_xlabel('X21 Label')
ax[1].set_ylabel('Y2 Label')
ax[1].set_title('Graph Title 2')
ax[1].text(4,2, r"$y=\sqrt[3]{x^2}$", fontsize=20, color="k")
```

### There are a whole list of things we can change and manually edit in our graphs. However, what we want to change is on a case by case basis and not worth teaching every last detail. Therefore, as you plot, its beneficial if you learn how to search online and most of the times a solution will be available for you online

<img src="../jupyter-notebook-images/plot-all-components.jpg" width="500">

Source: https://matplotlib.org/stable/tutorials/introductory/quick_start.html#sphx-glr-tutorials-introductory-quick-start-py

Here is a list of things we can change
* color = color of the line / markers
* marker = the symbol/shapes used for the markers
* markersize = size of the marker
* ls = line style
* lw = line width
* fontsize = changes the fontsize of the text
* markerfacecolor = the color of the marker
* markeredgecolor = the color of the edge of the marker

```python
fig, ax = plt.subplots(dpi = 75)
fig. set_size_inches(8,8)

ax.plot(x, y, color = 'r', ls ='--', marker='o', lw=3, markersize=12, markerfacecolor='none', markeredgecolor='k',  label = "Sample Line")
ax.set_xlabel('X Label', fontsize=20)
ax.set_ylabel('Y Label', fontsize=20)
ax.set_title('Graph Title 2', fontsize=30, fontweight ="bold")

ax.legend()
plt.grid(color = 'green', linestyle = '--', linewidth = 2)
```

# 7) Plotting Data From Lec 3 (Matplotlib)
Recall the code from Lec 3 - Importing and Manipulating Data where we took the averages for each of the three runs and combined them into a data table for each experiment.

```python
import pandas as pd
df2 = pd.read_excel('../Sample-Data/Sample-Data.xlsx', sheet_name="Agitation_and_Mixing")
mixing_data_avg = df2.groupby(['Impeller','Speed (RPM)']).mean()
mixing_data_avg.reset_index(inplace=True)
mixing_data_avg
```

### Now, we will plot this data using the skills learned in this lesson

First we will split the data into pitch data and turbine data by extract from df2 based on the impeller label

```python
pitch_data = mixing_data_avg[mixing_data_avg["Impeller"] == "Pitch"]
turbine_data = mixing_data_avg[mixing_data_avg["Impeller"] == "Turbine"]
```

Next, we will plot two scatter plots on top of each other

```python
plt.scatter(pitch_data["Speed (RPM)"],pitch_data["Avg Force (N)"], label="Pitch")
plt.scatter(turbine_data["Speed (RPM)"],turbine_data["Avg Force (N)"], label="Turbine")

plt.title("RPM vs Force")
plt.ylabel("Force (N)")
plt.xlabel("Rotations Per Minute (RPM)")
plt.legend()
```
